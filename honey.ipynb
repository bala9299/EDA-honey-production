### Draw a plot between the year and total productions.

```python
YEAR=df.groupby('year')
plt.plot(YEAR['productions'].sum(),marker='o')
plt.xlabel("Year")
plt.ylabel("productions")
plt.title("Total productions in dollars by year");
```

### Draw a horizontal barplot between the value of production and year.


```python
#total value of production by year
YEAR['value_of_prod'].sum().plot(kind='barh')
plt.xlabel('Total value of production')
plt.title("Total value of honey productions in dollars by year");
```

### List out the states between 2019 to 2021 in the order of productions.

```python
df[df['year'].between(left=2019,right=2021)].sort_values(by='productions',ascending=False)
```
